```python
import numpy as np
scores = np.load('results/P-1_scores.npy')
labels = np.load('results/P-1_labels.npy')

thr = np.percentile(scores, 98.5)
pred = (scores >= thr).astype(int)

# 이상 구간별 탐지 여부 확인
anomaly_regions = [[2149, 2349], [4536, 4844], [3539, 3779]]
for s, e in anomaly_regions:
    detected = pred[s:e].sum()
    print(f"구간 {s}~{e}: 탐지된 시점 {detected}개 / 전체 {e-s}개")
```

```python
import numpy as np
import pandas as pd

df = pd.read_csv('labeled_anomalies.csv')
df
```

```python
# P-1 채널 train/test shape 확인
train = np.load('train/M-6.npy')
test = np.load('test/M-6.npy')

print('train shape:', train.shape)
print('test shape:', test.shape)
print('train dtype:', train.dtype)
```

```python
pd.DataFrame(train).head(10)
```

```python
# 각 컬럼별로 0이 아닌 값이 몇 개인지 확인
non_zero = (pd.DataFrame(train) != 0).sum()
print(non_zero)
```

```python
# test 데이터도 확인
non_zero_test = (pd.DataFrame(test) != 0).sum()
print(non_zero_test)
```

```python
non_zero_test.sort_values(ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(5, 1, figsize=(15, 10))

for i, col in enumerate([0, 5, 33, 11]):
    axes[i].plot(train[:, col], label=f'train col {col}')
    axes[i].set_ylabel(f'col {col}')

plt.tight_layout()
plt.show()
```

```python
# test 데이터 + 이상 구간 표시
labels = np.zeros(8505)
for s, e in [[2149, 2349], [4536, 4844], [3539, 3779]]:
    labels[s:e] = 1

fig, axes = plt.subplots(3, 1, figsize=(15, 8))
for i, col in enumerate([0, 5, 6]):
    axes[i].plot(test[:, col], label=f'col {col}')
    axes[i].fill_between(range(8505), 0, 1,
                         where=labels==1,
                         alpha=0.3, color='red',
                         transform=axes[i].get_xaxis_transform())
    axes[i].set_ylabel(f'col {col}')

plt.tight_layout()
plt.show()
```
